# Line Plots And Subplot


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FormatStrFormatter
```

```python
df=pd.read_csv("immigrants_emigrants_by_sex.csv")
```

```python
df.head()
```

```python
table_immigrants = pd.pivot_table(data=df,index='District Name',values='Immigrants',aggfunc=np.sum)
table_immigrants
```

```python
table_emigrants = pd.pivot_table(data=df,index='District Name',values='Emigrants',aggfunc=np.sum)
table_emigrants
```

```python
table_i = pd.pivot_table(data=df,index='Year',values='Immigrants',aggfunc=np.sum)
table_i
```

```python
table_e = pd.pivot_table(data=df,index='Year',values='Emigrants',aggfunc=np.sum)
table_e
```

```python
plt.style.use('seaborn-darkgrid')
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(15,5)) 
ax[0].plot(table_i.index,table_i.Immigrants)
ax[0].xaxis.set_major_formatter(FormatStrFormatter('%d'))
ax[0].set_xlabel('year')
ax[0].set_ylabel('population')
ax[0].set_title('Immigrants')

ax[1].plot(table_e.index,table_e.Emigrants)
ax[1].xaxis.set_major_formatter(FormatStrFormatter('%d'))
ax[1].set_xlabel('year')
ax[1].set_ylabel('population')
ax[1].set_title('Emmigrants')
plt.show()
```

```python
plt.style.use('seaborn-bright')
plt.figure(figsize=(20,10))
plt.plot(table_immigrants.index,table_immigrants.Immigrants)
plt.plot(table_emigrants.index,table_emigrants.Emigrants)
plt.xlabel('District Name')
plt.ylabel('population')
plt.title('Immigrants and Emigrants')
plt.legend(["Immigrants","Emigrants"],loc=0)
plt.show()
```

# Scatter Plot

```python
plt.style.use('dark_background')
plt.figure(figsize=(10,5))
plt.scatter(table_immigrants.index,table_immigrants['Immigrants'],alpha= 0.7,color='red') 
plt.xlabel('District') 
plt.xticks(rotation=70)
plt.ylabel('population') 
plt.title('immigrants and Emigrants in that particular District') 
plt.scatter(table_emigrants.index,table_emigrants['Emigrants'],alpha= 0.7,color='green')
plt.legend(["immigrants","emmigrants"],loc=0)
plt.show()
```
