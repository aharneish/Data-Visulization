# Box Plot
taking the births dataset and plotting the box plot

```python
#importing the required packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_death=pd.read_csv("deaths.csv")
df_death.head()
```

```python
c_death={}
for i in df_death['Age'].unique():
    c_death[i]=df_death[df_death['Age']== i].Number.unique()
```

```python
plt.figure(figsize=(20,10))
plt.boxplot([x for x in c_death.values()], 
           labels = [x for x in c_death.keys()])
plt.xlabel('Age')
plt.ylabel('Number')
plt.title('Analysing the number of deaths in each Age group')
plt.savefig("1.BoxPlot.png",dpi=300)
plt.show()
```

```python
c_deaths={}
for i in df_death['District_Name'].unique():
    c_deaths[i]=df_death[df_death['District_Name']== i].Number.unique()
```

```python
plt.figure(figsize=(20,10))
plt.boxplot([x for x in c_deaths.values()], 
           labels = [x for x in c_deaths.keys()])
plt.xlabel('District_Name')
plt.ylabel('Number')
plt.title('Analysing the number of deaths in each district')
plt.savefig("2.BoxPlot2.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.hist(df_death['District_Name'],rwidth=0.9, alpha = 0.4, color= 'blue', bins = 15, edgecolor = 'red')
plt.xlabel('District_Name')
plt.ylabel('Frequency of Deaths')
plt.title('Inspecting the Death rate in the Districts')
plt.savefig("3.Histogram.png",dpi=300)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.hist(df_death['Age'],rwidth=0.9, alpha = 0.4, color= 'blue', bins = 15, edgecolor = 'red')
plt.xlabel('Age')
plt.ylabel('Frequency of Deaths')
plt.title('Inspecting the Death rate in the each Age Group')
plt.savefig("4.Histogram2.png",dpi=300)
plt.show()
```
